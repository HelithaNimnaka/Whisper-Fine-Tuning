```python
import whisper
from difflib import SequenceMatcher
from IPython.core.display import display, HTML  # Correct import

# Load Whisper model
model = whisper.load_model("small", device= 'cuda')  # Change model size if needed

# Define function for inference
def transcribe_audio(audio_path):
    result = model.transcribe(audio_path, language="si")  # Sinhala transcription
    return result["text"]

# Input audio file
audio_file = "/kaggle/input/duka-wadi-tharamata/Duka_Wadi_Tharamata.mp3"  # Change this to your actual file path

#  First Inference
print("\n Running First Inference...")
first_transcription = transcribe_audio(audio_file)
print("\n First Inference Result:\n", first_transcription)

#  Second Inference
print("\n Running Second Inference...")
second_transcription = transcribe_audio(audio_file)
print("\n Second Inference Result:\n", second_transcription)


def compare_transcriptions_html(text1, text2):
    words1 = text1.split()
    words2 = text2.split()

    matcher = SequenceMatcher(None, words1, words2)
    comparison = []

    for opcode, i1, i2, j1, j2 in matcher.get_opcodes():
        if opcode == 'equal':  
            comparison.extend([f"<span style='color:green; text-decoration: underline;'>{word}</span>" for word in words1[i1:i2]])
        elif opcode in ['replace', 'insert', 'delete']:  
            comparison.extend([f"<span style='color:red; text-decoration: line-through;'>{word}</span>" for word in words1[i1:i2]])

    return " ".join(comparison)


try:
    html_output = f"""
    <h3>üîç Detailed Comparison (Green Underline = Match, Red Strikethrough = Mismatch):</h3>
    <p style="font-size:16px;">{compare_transcriptions_html(first_transcription, second_transcription)}</p>
    """
    display(HTML(html_output))  # Fix: Ensure HTML is correctly rendered
except Exception as e:
    print("‚ö†Ô∏è HTML rendering failed. Showing raw output instead:\n")
    print(compare_transcriptions_html(first_transcription, second_transcription))
```
